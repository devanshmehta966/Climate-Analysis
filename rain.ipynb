```python
import pandas as pd
data=pd.read_excel('weatherAUS.xlsx')
data
```

```python
data.RainTomorrow.replace(('Yes','No'),(1,0),inplace=True)
data.RainToday.replace(('Yes','No'),(1,0),inplace=True)
data
```

```python
data=data.drop(['Date','Location','Evaporation','Sunshine','WindGustDir','WindDir9am','WindDir3pm'],axis=1)
data
```

```python
data1=data.fillna(0)
data1
```

```python
x=data1.iloc[:,:-1].values
x
```

```python
y=data[['RainTomorrow']]
y
```

```python
import numpy as np 
import pandas as pd
import statsmodels.api as sm
X = x
Y = y
model = sm.Logit(Y, X).fit()
predictions = model.predict(X)
model.summary()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)

from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
y_score_train = regressor.score(x_train,y_train)

print(y_score_train * 100)

regressor.fit(x_test,y_test)
y_score_test = regressor.score(x_test,y_test)

print(y_score_test * 100)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

y_train_pred=regressor.predict(x_train)
y_test_pred=regressor.predict(x_test)
```

```python
print(accuracy_score(y_train,y_train_pred))
print(accuracy_score(y_test,y_test_pred))
```

```python
print(confusion_matrix(y_train,y_train_pred))
print(confusion_matrix(y_test,y_test_pred))
```

```python
print(classification_report(y_train,y_train_pred))
print(classification_report(y_test,y_test_pred))
```
